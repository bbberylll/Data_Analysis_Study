```python
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt


# 모자이크 플롯 생성
mosaic(data, ['Gender', 'Sleep disorder'])

plt.title('Gender vs. Sleep disorder - Mosaic Plot')
plt.show()
```

```python
mosaic(data, ['Sleep disorder', 'Caffeine consumption'])

plt.title('Sleep disorder vs Caffeine consumption')
plt.show()
```

```python
mosaic(data, ['Sleep disorder', 'Wake up during night'])

plt.title('Sleep disorder vs  Wake up during night - Mosaic Plot')
plt.show()
```

```python
mosaic(data, ['Feel sleepy during day', 'Wake up during night'])

plt.title(' Feel sleepy during day vs Wake up during night')
plt.show()
```

```python
mosaic(data, ['Feel sleepy during day', 'Caffeine consumption'])

plt.title(' Feel sleepy during day vs Caffeine consumption')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ct = pd.crosstab(data['Gender'], data['Sleep disorder'])

sns.heatmap(ct, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap: Gender vs. Sleep disorder')
plt.xlabel('Sleep disorder')
plt.ylabel('Gender')
plt.show()
```

```python
ct2 = pd.crosstab(data['Sleep disorder'], data['Caffeine consumption'])

sns.heatmap(ct2, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap:  Caffeine consumption vs. Sleep disorder')
plt.xlabel('Sleep disorder')
plt.ylabel(' Caffeine consumption')
plt.show()
```

```python
ct3 = pd.crosstab(data['Wake up during night'], data['Caffeine consumption'])

sns.heatmap(ct3, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap:  Caffeine consumption vs. Wake up during night')
plt.xlabel('Wake up during night')
plt.ylabel(' Caffeine consumption')
plt.show()
```
